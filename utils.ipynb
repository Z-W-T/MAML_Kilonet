```python
import re
import math
```

```python
file = open('loss_MTF.txt')
loss = []
for line in file:
    if "origin" in line:
        continue
    print(line)
    pattern = r'Loss: [0-9\.]*(e-[0-9]*)?'
    matches = re.search(pattern, line)
    if matches:
        loss.append(float(matches[0][6:]))

average_loss = sum(loss)/len(loss)
average_loss
```

```python
MSE = 0.06866280734539032
PSNR = 10*math.log10(18.49/MSE)
PSNR
```

```python
bad_index = []
for i, l in enumerate(loss):
    psnr = 10*math.log10(1/l)
    if psnr < 30:
        bad_index.append(i)

print(bad_index)
```
